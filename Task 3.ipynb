# Data Visualisation

## Heatmap for getting the visual representation of correlation among features

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.heatmap(corr,annot=True,cmap='inferno')
plt.show()
```

## Drop all the non necessary columns

```python
df_new=df.drop(['Country','Region','City','State','Postal Code'],axis=1)
df_new.head()
```

## Histogram of Discount, Quantity, Sales & Profit

```python
df_new.hist(bins=40,figsize=(10,8))
plt.show()
```

## Count the number of records in each segment

```python
plt.figure(figsize=(8,5))
sns.countplot(df['Segment'],palette="Set1")
plt.show()
```

## Count the number of records in each region

```python
plt.figure(figsize=(8,5))
sns.countplot(df['Region'],palette="Set1")
plt.show()
```

## Count the number of records of each element of Sub-Category

```python
plt.figure(figsize=(10,6))
sns.countplot(df['Sub-Category'],palette="Set1")
plt.xticks(rotation=90)
plt.show()
```

## Count the number of record of each State

```python
plt.figure(figsize=(10,10))
sns.countplot(y=df['State'],palette="Set1")
plt.show()
```

## Visual Representation of a feature on another feature

```python
sns.pairplot(df,hue='Sub-Category')
plt.show()
```

## Percent Count of each item in Sub-Category

```python
plt.figure(figsize=(8,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Percent Count of each product')
plt.show()
```

## Percent of each element of Segment

```python
plt.figure(figsize=(10,8))
df['Segment'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Qantities ordered by each Segment')
plt.show()
```

## Profit vs Sales

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Sales'],df['Profit'],marker='o')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Profit vs Sales')
plt.show()
```

## Profit vs Discount

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Discount'],df['Profit'],marker='o')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Profit vs Discount')
plt.show()
```

## Quantity vs Profit

```python
ax1=sns.lineplot(df['Quantity'],df['Profit'],label='Quantity')
plt.show()
```

## Discount vs Profit

```python
sns.lineplot(df['Discount'],df['Profit'],label='profit')
plt.show()
```
